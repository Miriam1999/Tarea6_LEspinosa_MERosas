## Ejercicio 1

Recordar que:
$$X\sim binomial(n, p) \longrightarrow p_i = P(X = i) = \frac{n!}{i!(n - i)!}p^i(1 - p)^{n - i}, \quad i = 0,1,\cdots, n$$
Demostrar la validez de la siguiente ecuación:
>$$p_{i + 1} = \frac{n - i}{i+1} \frac{p}{1 - p} p_i \longrightarrow \text{Hablar de las ventajas de que sea recursiva}$$

### Solución de Lyha

### Solución de Miriam

#### Markdown

**Demostración matemática**:
1. Para demostrar la ecuación recursiva, se puede dividir $\frac{p_{i + 1}}{p_i}$ (en $p_{i + 1}$ se sustituye $i = i + 1$):
$$\frac{p_{i + 1}}{p_i} = \frac{\frac{n!}{(i + 1)!(n - (i + 1))!} p^{i + 1} (1 - p)^{n - (i + 1)}}{\frac{n!}{i! (n - i)!} p^i (1 - p)^{n - i}} = \frac{\frac{n! p^{i + 1} (1 - p)^{n - (i + 1)}}{(i + 1)!(n - (i + 1))!}}{\frac{n! p^i (1 - p)^{n - i}}{i! (n - i)!}} = \frac{n! p^{i + 1} (1 - p)^{n - (i + 1)} i! (n - i)!}{(i + 1)!(n - (i + 1))! n! p^i (1 - p)^{n - i}} = \frac{p^{i + 1} (1 - p)^{n - i - 1} i! (n - i)!}{(i + 1)! (n - i - 1)! p^i (1 - p)^{n - i}}$$


2. Pasando los exponentes del denominador hacia el numerador (se pasan con el signo contrario), se obtiene que:
$$\frac{p^{i + 1 - i} (1 - p)^{n - i - 1 - n + i} i! (n - i)!}{(i + 1)! (n - i - 1)!} = \frac{p^{1} (1 - p)^{- 1} i! (n - i)!}{(i + 1)! (n - i - 1)!} = \frac{p \cdot i! (n - i)!}{(i + 1)! (n - i - 1)! (1 - p)}$$


3. Usando una propiedad de los factoriales, la cual dice que $\frac{i!}{(i + 1)!} = \frac{1}{i + 1}$, se obtiene que:
$$\frac{p \cdot i! (n - i)!}{(i + 1)! (n - i - 1)! (1 - p)} = \frac{p \cdot 1 (n - i)!}{(i + 1) (n - i - 1)! (1 - p)} = \frac{p (n - i)!}{(i + 1) (n - i - 1)! (1 - p)}$$


4. Usando otra propiedad de los factoriales, la cual dice que $\frac{n!}{(n - 1)!} = n$ (para este caso, $n = n - i$, por lo tanto, la propiedad que aplica en realidad es $\frac{(n - i)!}{(n - i - 1)!} = n - i$), se obtiene que:
$$\frac{p (n - i)!}{(i + 1) (n - i - 1)! (1 - p)} = \frac{p (n - i)}{(i + 1) (1 - p)}$$


5. Finalmente, despejando $p_{i + 1}$ se obtiene que:
$$\frac{p_{i + 1}}{p_i} = \frac{p (n - i)}{(i + 1) (1 - p)}$$
$$p_{i + 1} = \frac{p (n - i) p_i}{(i + 1) (1 - p)}$$

Lo cual, en efecto, **desmuestra la validez de la ecuación recursiva** definida a partir de la fórmula para sacar probabilidades de una distribución binomial. Es más útil usar esta ecuación debido a que uno se ahorra el cálculo de los factoriales, pues se sabe que el factorial es una función que requiere de múltiples operaciones iterativas y por lo tanto su costo computacional es mucho mayor.

## Ejercicio 2

Seguir un procedimiento similar al mostrado cuando se generó una distribución binomial, pero en esta caso genere un código que genere variables aletorias Poisson cuya función de distribución de probabilidad esta dada por:
>$$P(k, \lambda) = \frac{e^{-\lambda}(\lambda)^k}{k!}$$
    
> Demuestre matemáticamente que 
> $$P(k + 1) = \frac{\lambda}{k + 1}P(k)$$
> y a partir de esta relación genere variables aletorias que distribuyen poisson usando el método de la transformada inversa.

### Solución de Miriam

#### Markdown

**Demostración matemática**:
1. Similar al ejercicio anterior, se puede demostrar la validez de la ecuación dividiendo $\frac{P(k + 1)}{P(k)}$ (en $P(k + 1)$ se sustituye $k = k + 1$):
$$\frac{P(k + 1)}{P(k)} = \frac{\frac{e^{-\lambda}(\lambda)^{k + 1}}{(k + 1)!}}{\frac{e^{-\lambda}(\lambda)^k}{k!}} = \frac{e^{-\lambda}(\lambda)^{k + 1} k!}{(k + 1)! e^{-\lambda}(\lambda)^k} = \frac{(\lambda)^{k + 1} k!}{(k + 1)! (\lambda)^k}$$


2. Pasando los exponentes del denominador hacia el numerador (se pasan con el signo contrario), se obtiene que:
$$\frac{(\lambda)^{k + 1} k!}{(k + 1)! (\lambda)^k} = \frac{(\lambda)^{k + 1 - k} k!}{(k + 1)!} = \frac{(\lambda)^{1} k!}{(k + 1)!} = \frac{\lambda k!}{(k + 1)!}$$


3. Usando de nuevo la propiedad utilizada en el paso 3 del ejercicio anterior ($\frac{k!}{(k + 1)!} = \frac{1}{k + 1}$), se obtiene que:
$$\frac{\lambda k!}{(k + 1)!} = \frac{\lambda \cdot 1}{k + 1} = \frac{\lambda}{k + 1}$$


4. Finalmente, despejando $P(k + 1)$ se obtiene que:
$$\frac{P(k + 1)}{P(k)} = \frac{\lambda}{k + 1}$$
$$P(k + 1) = \frac{\lambda P(k)}{k + 1}$$

Lo cual, en efecto también demuestra la validez de la ecuación dada, obtenida a partir de la función de distribución de probabilidad de Poisson.

**El siguiente código se genera usando la lógica que sigue**:
1. Para llenar el primer elemento de la distribución (el vector de las probabilidades), es necesario sustituir $k = 0$ en la fórmula de $P(k, \lambda)$:
$$\frac{e^{-\lambda}(\lambda)^k}{k!} = \frac{e^{-\lambda}(\lambda)^0}{0!} = \frac{e^{-\lambda} \cdot 1}{1} = e^{-\lambda}$$

2. Para llenar los demás elementos, basta con seguir la fórmula ya dada:
$$P(k + 1) = \frac{\lambda}{k + 1}P(k)$$

3. Después se genera la distribución aleatoria en base a esas probabilidades, calculando la frecuencia absoluta de cada una de las $x$. 

4. Finalmente se comparan dos histogramas: uno usando la distribución de Poisson generada en el procedimiento previo, y otro usando una distribución de Poisson generada con `numpy.random.poisson()`.

#### Código

```python
import numpy as np
from scipy import optimize
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
np.random.seed(123) # semilla
def prob_acum_dist_poisson(l: "término lambda", k: "número de términos k"):
    """Esta función regresa un vector de la probabilidad acumulada 'P(x)' con base en los valores lambda 'l' y la constante 'k' 
    de la distribución de Poisson"""
    probabilidades = np.zeros(k) # vector de ceros tamaño k
    probabilidades[0] = np.exp(-l) # llenar el primer elemento
    def llenar_vector(i):
        nonlocal probabilidades
        probabilidades[i + 1] = (l * probabilidades[i]) / (i + 1)
    [llenar_vector(i) for i in range(k - 1)] # llenar los demás elementos
    return np.cumsum(probabilidades)
def generar_dist_discreta(p_acum: "vector de la probabilidad acumulada de la distribución a generar", 
                          k: "cantidad de números aleatorios a generar"):
    """Esta función regresa el vector con la frecuencia de cada uno de los números aleatorios generados"""
    x = np.random.rand(k) # generar N valores aleatorios uniformemente distribuidos entre 0 y 1
    y = pd.Series([sum([1 for p in p_acum if p < xi]) for xi in x]) # calcular la frecuencia de dichos de valores
    return y
def graficar_hist_discreto(distribucion: "distribución a graficar el histograma", title: "título del histograma"):
    """Esta función plotea el histograma de una variable discreta dada una distribución (un vector con la frecuencia de cada
    uno de los valores aleatorios generados), y un título de la gráfica"""
    plt.figure(figsize = (15, 7))
    y, x = np.histogram(distribucion, bins = len(set(distribucion)), density = True) 
    plt.bar(list(set(distribucion)), y, color = "y")
    plt.xlabel("Valores aleatorios", fontsize = 12)
    plt.ylabel("Frecuencia relativa (probabilidad)", fontsize = 12)
    plt.title(title, fontsize = 20)
    plt.grid(color = "#95a5a6", linestyle='--', linewidth = 1.5, axis = "y", alpha = 0.7)
    plt.show()
k, l = 10000, 5 # número de muestras y lambda 'l'
p_acum = prob_acum_dist_poisson(l, k) # calcular la probabilidad acumulada con los lamnda 'l' y 'k' dados

# Usando el método de la transformada inversa
d_poisson_transformada = generar_dist_discreta(p_acum, k)
graficar_hist_discreto(d_poisson_transformada, "Histograma de distribución de Poisson con el método de la transformada inversa")

# Usando numpy
d_poisson_np = np.random.poisson(l, k)
graficar_hist_discreto(d_poisson_np, "Histograma de la distribución de Poisson usando NumPy")
```

```python
np.random.seed(123) # semilla
def var_acep_rechazo(R2: "variables distribuidas U ~ U(0, 1)'", R1: "variables distribuidas como g(x)", 
                     f: "función objetivo a generar", t: "función que mayora a f(x)"):
    """Esta función, en lugar de graficar el método de aceptación y rechazo, sirve para generar valores aleatorios que sean
    aceptados."""
    fx = f(R1)
    tx = t(R1)
    condition = R2 * tx <= fx # si se cumple la condición, los valores son aceptados
    return [R1[i] for i in range(len(R1)) if condition[i]] # se agregan a esta lista los valores que son aceptados
N = 500 # número de muestras
f = lambda x: 30 * (x**2 - 2 * x**3 + x**4) # f(x)
max_f = f(optimize.fmin(lambda x: -f(x), 0, disp = False)) # parámetro 'a'
t_cte = lambda x: max_f * np.ones(len(x)) # t constante
t_sin = lambda x: max_f * np.sin(np.pi * x) # t senoidal
R2 = np.random.rand(N) # valores aleatorios uniformemente distribuidos
R1 = np.random.rand(N) # valores aleatorios uniformemente distribuidos
grafica_acep_rechazo(R2, R1, f, t_cte, N) # primero ploteo la gráfica
print("El " + str((len(var_acep_rechazo(R2, R1, f, t_cte)) / N) * 100) + "% de los valores son aceptados con la t(x) constante."
     )
print("El " + str((len(var_acep_rechazo(R2, R1, f, t_sin)) / N) * 100) + "% de los valores son aceptados con la t(x) sinoidal.")
```
